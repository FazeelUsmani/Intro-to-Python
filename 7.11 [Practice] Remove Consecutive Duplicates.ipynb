```python
str = input()
str1 = []
count = 0
for i in range(0, len(str)-1):
    
    if str[i] == str[i+1]:
        count += 1
    elif str[i] != str[i+1] and count > 0:
        str1.append(str[i])
        count = 0
    else:
        str1.append(str[i])
str1.append(str[i+1])
    
for ele in str1:
    print(ele, end = '')
```
