### Merge two sorted arrays

```python
m = int(input())
arr1 = [int(x) for x in input().split() [:m]]
n = int(input())
arr2 = [int(x) for x in input().split() [:n]]

arr3 = []
i = 0
j = 0
while i < len(arr1) and j < len(arr2):
    if arr1[i] <= arr2[j]:
        arr3.append(arr1[i])
        #print("i ", arr3)
        i = i + 1
    elif arr1[i] > arr2[j]:
        arr3.append(arr2[j])
        #print("j ",arr3)
        j = j + 1
        
if i < len(arr1):
    for l in arr1[i : len(arr1)]:
        arr3.append(l)
else:
    for k in arr2[j :  len(arr2)]:
        arr3.append(k)
        
for ele in arr3:
    print(ele, end = ' ')
```
